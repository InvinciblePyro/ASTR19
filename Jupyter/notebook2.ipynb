```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)

sinx = np.sin(x)
cosx = np.cos(x)
sin4x = np.sin(4*x)
cos4x = np.cos(4*x)
```

```python
fig, axs = plt.subplots(nrows= 2, ncols= 2)

#first panel
axs[0, 0].plot(x, sinx)
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("sin(x)")
axs[0, 0].set_title(r'$\sin(x)$')

#second panel
axs[0, 1].plot(x, cosx)
axs[0, 1].set_xlabel("x")
axs[0, 1].set_ylabel("cos(x)")
axs[0, 1].set_title(r'$\cos(x)$')

#third  panel
axs[1, 0].plot(x, sin4x)
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("sin(x)*4")
axs[1, 0].set_title(r'$\sin(x)$')

#second panel
axs[1, 1].plot(x, cosx)
axs[1, 1].set_xlabel("x")
axs[1, 1].set_ylabel("cos(x)*4")
axs[1, 1].set_title(r'$\cos(x)$')

fig.tight_layout()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(6, 6))


#first panel
axs[0].plot(x, sinx, label=r'$\sin(x)$')
axs[0].plot(x, sin4x, label=r'$\sin(4x)$')
axs[0].set_xlabel("x")
axs[0].set_ylabel("y(x)")
axs[0].legend(fontsize=13)
axs[0].set_aspect(np.pi)

#second panel
axs[1].plot(x, cosx, label=r'$\cos(x)$')
axs[1].plot(x, cos4x, label=r'$\cos(4x)$')
axs[1].set_xlabel("x")
axs[1].set_ylabel("y(x)")
axs[1].legend(fontsize=13)
axs[1].set_aspect(np.pi)

fig.tight_layout()
```
