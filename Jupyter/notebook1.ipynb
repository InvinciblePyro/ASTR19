```python
import numpy as np
from astropy.table import Table

Gaia_table = Table.read("Gaia_1kpc.fits", format="fits")

parallax = Gaia_table["parallax"]
parallax_arc = parallax / 1000
g_app_mag = Gaia_table['phot_g_mean_mag']
```

```python
def get_dist(p_angle): return (1/p_angle)
def get_absolute_mag(D, app_mag): return (app_mag - 5  * np.log10(D/10))

dist = get_dist(parallax_arc)
abs_mag = get_absolute_mag(dist, g_app_mag)

distances = get_dist(parallax_arc)

g_abs_mag = g_app_mag - 5  * np.log10(dist/10)

```

```python
import matplotlib.pyplot as plt

# sets the default parameters for all graphs
plt.rcParams['font.family'] = "Times New Roman"
plt.rc('font', size = 18)

t = plt.hist(distances, bins = 100)
plt.xlabel("Distance [pc]", fontsize = 20)
plt.ylabel("Number of stars", fontsize = 20)
```
